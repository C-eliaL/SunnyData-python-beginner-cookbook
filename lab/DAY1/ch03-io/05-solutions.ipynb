### 重點提醒

**input() 的三個重點**：
1. 永遠回傳 str
2. 需要手動轉型（int, float）
3. 提示訊息是選用參數

**print() 的重要參數**：
- `sep`: 分隔符（預設為空格）
- `end`: 結尾字元（預設為換行）

**f-string 常用格式**：
- `:.2f` - 浮點數 2 位小數
- `:,` - 千分位分隔符
- `:>10` - 右對齊寬度 10
- `:<10` - 左對齊寬度 10
- `:^10` - 置中對齊寬度 10
- `:05d` - 整數寬度 5 補零
- `:.1%` - 百分比 1 位小數
